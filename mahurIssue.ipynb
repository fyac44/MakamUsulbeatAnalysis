# ISSUES WITH MAHUR MAKAM

As we were analyzing the scores we get some "suspicious" results. The most noticeable was with the mahur makan, where the "mahur" note never appeared. We selected some files and retrieved all the notes and printed their full name in order of appearance.


Then we check the original xml file (the one with the key signatures) and the PDF file on the database. We found that in the PDF file the mahur note sometimes is represented as a F-sharp 5 and not as a F-slash-quarter-sharp 5.

## First Example:
mahur--sarki--aksak--acirim_asik--giriftzen_asim_bey

To analyse this file with music21 we had to remove the key signature and save the new xml:

```python
from score_handler import *

file = './selected_scores/mahur--sarki--aksak--acirim_asik--giriftzen_asim_beynoKeySig.xml'
notes_on_file(file)
```

Notice that the second note is F-slash-quarter-sharp5 "Mahur" in the PDF but read as F5 "Acem" by music21.

## Second example:
mahur--sarki--aksak--iki_gozum--kemal_emin_bara

```python
file = './selected_scores/mahur--sarki--aksak--iki_gozum--kemal_emin_baranoKeySig.xml'
notes_on_file(file)
```

In this case, the PDF file always replace the accidental "slash-half-sharp" accidental shown in the key for a "sharp" or a "natural" accidental. It's the same in the original xml file.


Also, there are incosistencies between the score and the notes obtained by music21.

This issue could had a very important impact in our results and in our conclusions if we haven't noticed them before. Maybe the dataset must be carefully reviewed before use
